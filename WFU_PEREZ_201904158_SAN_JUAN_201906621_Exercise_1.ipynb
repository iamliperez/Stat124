## **Item 1A**

To construct the stem-and-leaf display (SALD), we first compute the ratio `R/L`.

**Note:** We use the **Dixon and Kronmal Formula** to compute the `L`.

```r
R <- max(weekly.salaries) - min(weekly.salaries)
L <- floor(10*log(23, 10))
R/L
```

From the previous results, since the closest to `R/L = 33.92` is 50, we will utilize stretched SALD.

**Stretched SALD of the**

**Weekly Salaries (in US$) of 23 WB Country Managers**

```r
stem.leaf(weekly.salaries, m=2)
```

### **Interpretation:**


> The stem-and-leaf display of the weekly salaries (in US$) of 23 WB country managers across the globe shows that most of the values from the batch are located on the higher amounts of salary (i.e., stems 8 and 9), and it follows that the distribution of the data seems to be negatively skewed. Moreover, the median can be found in the stem 8* where there is also high concentration of the data. With this, we can infer that the salaries of the world bank country managers across the globe are relatively high.

## **Item 1B**

We use the `hist()`function to create the histogram of the weekly salaries (in US$) of 23 WB country managers across the globe.

**Note:** `breaks = 13` from **Item 1A** (`L = 13`).

```r
hist(weekly.salaries,
      breaks = 13,
      main = "Histogram of the Weekly Salaries (in US$) 
      of 23 WB Country Managers",
      xlab = "Amount of Salary"
      )
```

### **Comparison and Explanation:**
> The SALD contains more information whereas we can see the exact values of the batch, compared to the histogram which only gives us the summarized information. Both SALD and histogram demonstrate similar characteristics in terms of concentration of values and skewness, that is, most of the values are located in the higher amount of salary and they are both seemed to be negatively skewed.

## **Item 1C**

We use the `boxplot()`function to create the boxplot of the weekly salaries (in US$) of 23 WB country managers across the globe.

```r
boxplot(weekly.salaries, 
        main = "Boxplot of the Weekly Salaries (in US$) of 23 WB Country Managers",
        xlab = "Amount of Salary",
        horizontal = TRUE)
```

### **Interpretation:**

> Looking at the median and the fourths, we can see that the median line is closer to the lower fourth which implies that the distribution is positively skewed. However, observe that the position of the box itself is closer to the right whisker and we can also see that the tail length of the left whisker is longer, that intuitively implies that the distribution seems to be negatively skewed which is consistent to the previous results from SALD and histogram. Lastly, as presented in the boxplot, we do not have any outlying data points.

## **Item 1D**

We use the `plot()`function to create the quantile plot of the weekly salaries (in US$) of 23 WB country managers across the globe.

```r
n <- length(weekly.salaries)

plot(((1:n)-0.5)/n, sort(weekly.salaries, decreasing = FALSE),
      main = "Quantile Plot of the Weekly Salaries (in US$) of 
      23 WB Country Managers",
      xlab = "Fraction of Data",
      ylab = "Quantiles of the Amount of Salary")
```

### **Shape of the Underlying Distribution:**



> Based from the quantile plot, it is shown that the distribution exhibits non-normality since the plot does not follow a straight line. In addition, notice that there are observable curves which might indicate that the distribution is not symmetric.


We also utilize the `stripchart()` function to check the shape of the underlying distribution.




```r
stripchart(weekly.salaries,
            vertical = FALSE,
            pch = 21,
            cex = 2,
            bg = "blue",
            main = "Plot of the Weekly Salary")
```

> Consistent with what is shown in the quantile plot, the distribution is not symmetric and the created stripchart shows that the underlying distribution seems to be negatively skewed.


## **Item 1E**



> We use the `quantile()` function to determine the median, 5th, and 95th percentile, wherein the **median is 830 USD**, the **5th percentile is 616 USD**, and the **95th percentile is 989.8 USD**. 



```r
quantile(weekly.salaries,
          na.rm = T,
          probs = c(0.05,0.50,0.95))
```

# **Item 2 | Boxplot of the Amounts of Rainfall from Seeded Clouds**

We first load the dataset from **Rainfall from Cloud-Seeding.xlsx** and name it as `rainfall`.

```r
rainfall <- read_excel("/content/Rainfall from Cloud-Seeding.xlsx", sheet = 1)
```

We use the `boxplot()` function to create the boxplot of the amounts of rainfall from clouds seeded with silver oxide.

```r
boxplot(rainfall$seeded,
        main = "Boxplot of the Amounts of Rainfall from Seeded Clouds",
        xlab = "Rainfall in Acre-Feet",
        horizontal = TRUE)
```

### **Why are there so many outliers?**

> To answer this question, it is important for us to know what is cloud seeding. According to the Water Corporation (n.d.), cloud seeding is the process of implanting clouds with particles, such as silver iodide crystals, to artificially generete rain. With this, one of the main reasons of having many outliers may be brought by the nature of the process. Selecting sample is randomized in a way that the characteristics of the selected clouds may have huge variability due to its natural distribution. Considering this, there is a chance that the selected clouds may already contain high volume of rain resulting to a much higher amount of rainfall after implanting the silver oxide particles.
